# Perfect Pace Linear Programming Approach

```python
import pandas as pd
import numpy as np
import cvxpy as cp
import matplotlib.pyplot as plt
from datetime import datetime
import math
import gpx_parser
```

# Utility Functions

```python
def total_distance(segments):
    """
    Returns the sum of the distances of each segment in segments.
    """
    result = 0
    for segment in segments:
        result += segment.distance
    return result

def total_elevation(segments):
    """
    Returns the net change in elevation from the beginning of the first segment to the end of the last segment.
    """
    result = 0
    for segment in segments:
        result += segment.elevation_change
    return result

def meters_to_feet(meters):
    """
    Converts distance in meters to feet.
    """
    return meters * 3.28084

def meters_to_miles(meters):
    """
    Converts distance in meters to miles.
    """
    return meters / 1609.34

def miles_to_meters(miles):
    """
    Converts distance in miles to meters.
    """
    return miles * 1609.34

def to_min_per_mile(seconds_per_meter):
    """
    Converts paces in seconds per meter to mins per mile.
    """
    return (seconds_per_meter * 1609.34) / 60

def decimal_to_min_per_mile(decimal):
    """
    Takes decimal min per mile and converts to time representation as a string.
    """
    minutes = int(decimal)
    remaining = decimal-minutes
    seconds = int(60*remaining)
    return f'{minutes}:0{seconds}' if seconds < 10 else f'{minutes}:{seconds}'
    
def get_distances(segments):
    """
    Returns an array of distances marking the distance in miles of the start of each segment.
    """
    start_distance = 0
    distances = []
    for segment in segments:
        start_distance += meters_to_miles(segment.distance)
        distances.append(start_distance)
    return distances
    
def get_elevations(segments):
    """
    Returns an array of elevations marking the elevation in feet of the start of each segment.
    """
    start_elevation = meters_to_feet(segments[0].start_ele)
    elevations = []
    for segment in segments:
        start_elevation += meters_to_feet(segment.elevation_change)
        elevations.append(start_elevation)
    return elevations

def get_grade_adjusted_paces(segments, base_pace):
    """
    Returns an array of paces marking the grade adjusted pace to run in each segment in mins per mile.
    """
    grade_adjusted_paces = []
    for segment in segments:
        grade = (segment.elevation_change/segment.distance) * 100 # (meters/meters) * percent = percent
        pace = base_pace + grade_adjustment(grade)
        grade_adjusted_paces.append(pace / 60)   # Convert seconds per mile to mins per mile: seconds * (1 min / 60 seconds) = mins
    return grade_adjusted_paces
```

# Calculate Grade Adjusted Paces
For a segment i, the grade adjusted pace can be defined as follows:
$$gap_i = bp + \text{adj}(g_i)$$
Where:
* $gap_i$ = Grade adjusted pace for segment i
* $bp$ = Base pace
* adj($g_i$) = Grade adjustment function defined by Jack Daniel's running calculator
* $g_i$ = Elevation grade of segment i

For all $i = 1 ... n-1$:
$$gt = \sum_{i=0}^{n-1} (bp + \text{adj}(g_i)) * d_i$$
Where:
* $gt$ = Goal time
* $d_i$ = Distance of segment i  
This can be rearranged as follows:
$$gt = \sum_{i=0}^{n-1} bp*d_i + \sum_{i=0}^{n-1} \text{adj}(g_i) * d_i$$
$$gt = bp * \sum_{i=0}^{n-1} d_i + \sum_{i=0}^{n-1} \text{adj}(g_i) * d_i$$
$$bp = \frac{gt - \sum_{i=0}^{n-1} \text{adj}(g_i) * d_i}{\sum_{i=0}^{n-1} d_i}$$  
$$bp = \frac{\text{Goal Time} - \text{Total Grade Adjusted Time}}{\text{Total Distance}}$$

```python
def grade_adjustment(grade):
    """
    Calculates the pace adjustment for a given segment based on that segment's elevation grade. 
    This is adj() in above description.
    
    Per Jack Daniel's Formula:
    1% elevation increase => +12sec/mile
    1% elevation decrease => -7sec/mile
    
    The unit of the final output should be seconds per mile.
    """
    if grade < 0:
        return 7 * grade
    else:
        return 12 * grade

def find_time_adjustment(segments):
    """
    Calculates the total time adjustment for all segments.
    
    i.e. \sum_{i=0}^{n-1} adjustment(grade_i) * distance_i
    
    Where distance is in miles and time is in seconds. The unit of the final output should be seconds.
    """
    result = 0
    for segment in segments:
        distance = meters_to_miles(segment.distance)    # Distance in miles
        grade = (segment.elevation_change/segment.distance) * 100 # (meters/meters) * percent = percent
        result += grade_adjustment(grade) * distance    # seconds/mile * miles = seconds
    return result
    
def find_base_pace(segments, goal_time):
    """
    Calculates the base pace that fits the goal time with pace adjustments for elevation.
    
    i.e. base pace = (goal time - time adjustment) / total distance
    
    Where goal time is in seconds, time adjustment is in seconds, and distance is in miles. 
    The unit of the final output should be seconds per mile
    """
    grade_adjusted_time = goal_time - find_time_adjustment(segments)    # Time in seconds
    distance = meters_to_miles(total_distance(segments))    # Distance in miles
    return grade_adjusted_time / distance
```

```python
def plot_grade_adjusted_paces(segments, goal_time, loop_name = None):
    """
    Creates an elevation plot with grade adjusted paces overlayed for a given loop (defined by segments) and goal time.
    """
    base_pace = find_base_pace(segments, goal_time)
    paces = get_grade_adjusted_paces(segments, base_pace)
    plot_paces(segments, paces, loop_name)

def plot_paces(segments, paces, loop_name = None):
    """
    Creates an elevation plot with paces overlayed for a given loop (defined by segments).
    """
    distance = meters_to_miles(total_distance(segments))
    distances = get_distances(segments)
    elevations = get_elevations(segments)

    fig, ax1 = plt.subplots(figsize=(12, 6))

    ax1.plot(distances, elevations, '-', color='b', label='Elevation')
    ax1.set_xlabel('Distance (Miles)')
    ax1.set_ylabel('Elevation (Feet)', color='b')
    ax1.set_xlim(0, distance)  # Set x-axis limit (distance)
    ax1.set_ylim(min(elevations) - 20, max(elevations) + 20)  # Set y-axis limit (elevation)
    ax1.tick_params(axis='y', labelcolor='b')

    # Create a secondary y-axis for the paces
    ax2 = ax1.twinx()

    # Plot the paces on the secondary y-axis (ax2)
    ax2.plot(distances, paces, '--', color='r', label='Pace')
    ax2.set_ylabel('Pace (mins per mile)', color='r')
    ax2.set_ylim(min(paces) - .25, max(paces) + .25)  # Adjust the y-axis limit for paces
    ax2.tick_params(axis='y', labelcolor='r')

    # Add a title
    if loop_name:
        plt.title(f'Elevation Profile with Paces for {loop_name}')
    else:
        plt.title('Elevation Profile with Paces')
    fig.legend(loc="upper right", bbox_to_anchor=(0.945, 0.9375))

    # Show the plot
    fig.tight_layout()  # Adjust layout to prevent overlap
    plt.show()
```

```python
segments = gpx_parser.parse_gpx('data\Lakefront-Loops-5K.gpx')
plot_grade_adjusted_paces(segments, 20*60, loop_name = 'Lakefront-Loops-5K')
```

# Linear Programming Algorithm
Objective: 
* minimize the absolute (or squared) difference between the grade adjusted pace of each segment and the optimal pace
* minimize the magnitude of pace change
* maximize the distance between different pace changes

Constraints: 
* Have at most k different pace segments throughout the entire loop
* Total time of paces should equal the goal time

### Mathematical Formulation:
$$\text{minimize } \sum_{i}^{n-1} |p_i - \text{gap}_i| \text{ or minimize }\sum_{i}^{n-1} (p_i - \text{gap}_i)^2$$
$$\text{s.t.} \sum_{i}^{n-1} d_i \cdot p_i = gt$$
$$M \cdot z_i \geq p_{i+1} - p_i \quad i = 0...n-2$$
$$M \cdot z_i \geq p_{i} - p_{i+1} \quad i = 0...n-2$$

$$\sum_{i}^{n-2} z_i \leq k$$
$$z_i \in {0, 1}, p_i \geq 0$$

Where:
* $z_i$ is a binary variable which is 0 when segment i has the same pace as segment i+1 and 0 otherwise.
* $p_i$ is the optimal pace of segment i
* $M$ is a large constant
* $k$ is the maximum number of pace changes  
All other variables have been previously defined

For absolute difference linearization:
$$\text{maximize } \sum_{i} t_i$$
$$t_i \geq p_i - \text{gap}_i$$
$$t_i \geq \text{gap}_i - p_i$$

```python
def optimize_pace(segments, total_time = 0, pace_adjustments = 3, lmbda = .1, M = 10, eps = 1):
    """
    Return optimal pace segments for given segments of a .gpx file. Paces should be in mins per mile.
    """
    n = len(segments)
    distance = meters_to_miles(total_distance(segments))
    bp = find_base_pace(segments, total_time)
    gaps = get_grade_adjusted_paces(segments, bp)

    # Variables
    p = cp.Variable(n, integer=False)  # Create a vector of variables for the paces of each segment
    z = cp.Variable(n-1, integer=True) # Create z variables for the first N-1 segments
    t = cp.Variable(n, integer=False)

    # Define constraints
    constraints = []
    variable_constraints = [p >= 0, z >= 0, z <= 1]
    total_time_constraint = [total_time == sum([meters_to_miles(segments[i].distance) * p[i] for i in range(n)])]    
    pace_change_constraints = [z[i] >= p[i+1] - p[i] for i in range(n-1)]
    pace_change_constraints += [z[i] >= p[i] - p[i+1] for i in range(n-1)]    
    pace_change_limit = [sum(z) == pace_adjustments]
    absolute_linearization = [t[i] >= p[i] - gaps[i] for i in range(n)]
    absolute_linearization += [t[i] >= gaps[i] - p[i] for i in range(n)]
    constraints += variable_constraints + total_time_constraint + pace_change_constraints + pace_change_limit + absolute_linearization
    
    # Define objective function
    # objective = cp.Minimize(sum([(gaps[i] - p[i])**2 for i in range(n)])) # Squared Loss
    objective = cp.Minimize(sum(t)) # Absolute Loss
    
    # Define and solve problem
    problem = cp.Problem(objective, constraints)
    problem.solve(solver=cp.GUROBI, verbose=False)

    # Construct optimal pace segments
    if problem.status in [cp.OPTIMAL, cp.OPTIMAL_INACCURATE]:
        paces = p.value
        pace_changes = z.value
    else:
        # If optimization infeasible throw error
        raise Exception(f"Optimization failed with status {problem.status}")

    return paces/60, pace_changes  # Paces in mins per mile
```

```python
goal_time = 60*20
segments = gpx_parser.parse_gpx('data\Lakefront-Loops-5K.gpx')
paces, pace_changes = optimize_pace(segments, total_time = goal_time, pace_adjustments = 5)
plot_paces(segments, paces, loop_name = 'Lakefront-Loops-5K')
```
