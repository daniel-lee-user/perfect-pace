```python
import numpy as np
import cvxpy as cp
import racecourse
import os
import matplotlib.pyplot as plt
from pacing_plan import PacingPlan
```

```python
class PacingPlanLP(PacingPlan):
    def __init__(self, race_course, target_time, total_paces):
        super().__init__(race_course, target_time, total_paces)
        
    def formulate_lp_problem(self):
        n_segments = self.get_n_segments()
        paces = cp.Variable(n_segments)
        changes = cp.Variable(n_segments-1, integer=True)
        absolutes = cp.Variable(n_segments)
        
        adjustments = self.get_pace_adjustments(self.race_course.grades)
        objective = cp.Minimize(cp.sum(absolutes))

        constraints = [
           paces @ self.get_distances() == self.target_time,
            paces >= 0,
            changes >= 0,
            changes <= 1,
            cp.sum(changes) == self.current_m_paces - 1,
            absolutes >= paces - (self.base_pace + adjustments),
            absolutes >= (self.base_pace + adjustments) - paces
        ]
        constraints += [changes[i] >= paces[i+1] - paces[i] for i in range(n_segments-1)]
        constraints += [changes[i] >= paces[i] - paces[i+1] for i in range(n_segments-1)]
        
        return cp.Problem(objective, constraints)

    def solve_lp_problem(self):
        problem = self.formulate_lp_problem()
        problem.solve(solver=cp.GUROBI)
        if problem.status == cp.OPTIMAL:
            self.true_paces_full = problem.variables()[1].value
        else:
            raise ValueError("LP problem is infeasible")

    def calculate_recommendations(self):
        self.solve_lp_problem()
        self.full_seg_times = np.multiply(self.get_distances(), self.true_paces_full)
```

```python
file_path = 'data\Lakefront-Loops-5K.gpx'
course_name = os.path.basename(file_path).split('.')[0]
course = racecourse.RealRaceCourse(course_name, file_path)
```

```python
lp_plan = PacingPlanLP(course, 20, 5)
```

```python
lp_plan.solve_lp_problem()
```

```python
paces = lp_plan.true_paces_full
mask = np.diff(paces) != 0
indices = np.where(mask)[0] + 1
```
